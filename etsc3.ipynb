### Ex 2: Integrate f(x) between limits, $\displaystyle \int_a^b f(x) dx$

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.display import Image

from scipy import *
import scipy.linalg as la

from scipy.integrate import quad, dblquad, tplquad #quadrature, doublt quadrature, triple quadrature


def f(x):
  a=1;
  return a* x**2
x_lower=0*(-np.pi); x_upper = np.pi; # limits
val, abserr = quad(f,x_lower, x_upper)  #or give valuues straight away
print ("integral value=",val,",absolute error = ", abserr)


plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"]= True

n=256
X= np.linspace(x_lower, x_upper,n, endpoint=True)
Y= f(X)

#fig, ax= plt.subplots ()
#ax.plot(X, Y)

fig, ax= plt.subplots ()
for n in range(256):
  ax.plot(X, Y, 'm-.', linewidth='5', )  #or plt.plot(X,Y, '--', 'color=r')
  ax.fill_between(X, Y, 0, color='blue', alpha=1)



```
